<div class="content">
    <h2>Problem Statement</h2>
    <p>
        We need to build an artificial neural network (ANN) to select the best basketball team based on player data. The goal is to find the "optimal team" of 5 players by analyzing different player characteristics from a dataset of NBA players.
    </p>
    <p>
        Our approach to selecting the optimal basketball team is based on projecting the team's win percentage. To do this, we grouped the top 5 players' stats from a team (such as points per game, assists per game, etc.) to create a set of team-level statistics.
    </p>
    <p>
        We then used these team statistics as inputs to train our model, with the team's actual win percentage serving as the label (output). The model learns the relationship between team stats and win percentage, allowing it to predict a win percentage for any new team based on its player stats.
    </p>
    <p>
        By passing team data through the trained model, we can project the win percentage and use this to identify the "optimal" team.
    </p>
</div>

<div class="algorithm-block">
    <p><strong>Artificial Neural Network (ANN) Algorithm</strong></p>
    <ol>
        <li>Initialize the weights <code>w<sub>i</sub></code> and biases <code>b<sub>i</sub></code> for all neurons with small random values. This ensures that the network does not start with any inherent bias.</li>
        <li>For each training example in the dataset:
            <ul>
                <li>Feed the input values <code>x<sub>1</sub>, x<sub>2</sub>, ..., x<sub>n</sub></code> into the network.</li>
            </ul>
        </li>
        <li><strong>Forward Propagation:</strong> Calculate the output of the network by propagating the input values through each layer:
            <ul>
                <li>For each neuron, compute the weighted sum of inputs: <code>z = w<sub>1</sub>x<sub>1</sub> + w<sub>2</sub>x<sub>2</sub> + ... + b</code>.</li>
                <li>Apply an activation function (e.g., sigmoid or ReLU) to the weighted sum to produce the output for each neuron.</li>
            </ul>
        </li>
        <li>Compute the error at the output layer by comparing the predicted output <code>y^</code> with the actual target value <code>y</code> using a cost function (e.g., mean squared error).</li>
        <li><strong>Backpropagation:</strong> Update the weights and biases by calculating the gradients of the cost function with respect to the weights and biases:
            <ul>
                <li>Compute the partial derivatives of the cost function with respect to the weights and biases at the output layer.</li>
                <li>Propagate these gradients backward through the hidden layers, adjusting the weights and biases in each layer using gradient descent.</li>
            </ul>
            <p style="text-align: center;">
                <code>w<sub>i</sub> ← w<sub>i</sub> - η ∂C/∂w<sub>i</sub></code>  
                <br>
                <code>b<sub>i</sub> ← b<sub>i</sub> - η ∂C/∂b<sub>i</sub></code>
            </p>
            <p>where <code>η</code> is the learning rate and <code>C</code> is the cost function.</p>
        </li>
        <li>Repeat steps 2 to 5 for multiple epochs, passing the entire training dataset through the network, until the error is minimized and the network's performance is satisfactory.</li>
        <li>Use the trained model to make predictions on new data by performing forward propagation and generating output values.</li>
    </ol>
</div>

    <p>
        We also added in win rates to each of our players so that when we are making teams later, we would have that players team's win percentage on them so that we can train our model to predict win percentages.
    </p>
</div>

```python
def winRateFromYear(year):

    winRateDf = pd.read_html(f'https://www.teamrankings.com/nba/stat/win-pct-all-games?date={year}-06-16')[0]

    winRateDf['Win PCT']= winRateDf[f'{year - 1}'] 

    winRateDf['season'] = year

    winRateDf = winRateDf[['Team', 'Win PCT', 'season']]

    return winRateDf
```

```python
def getWinRates():
    # merge the two dataframes on team with df and teams on team where season is 2004
    winRateDf = pd.DataFrame()

    for year in range(2004, 2024):
        winRateDf = pd.concat([winRateDf, winRateFromYear(year)], ignore_index=True)

    winRateDf.tail()

    return winRateDf

# If you don't want to run the web scraping code, set run to False
run = False

if run:
    winRateDf = getWinRates()
    winRateDf.to_csv('winRate.csv', index=False)
else:
    winRateDf = pd.read_csv('winRate.csv')
```

```python
updatedPlayerData = pd.merge(playerData, winRateDf, left_on=['team', 'season'], right_on=['Team', 'season'])

updatedPlayerData.drop(['Team'], axis=1, inplace=True)

updatedPlayerData.head()
```

<div class="content">
    <p>
        Since the goal of our project is only to predict the optimal team consisting of 5 players, we had to shrink the roster size for each team down to 5, but we didn't want to do that randomly, instead we just wanted the starting 5 players from each team to be kept in our data. For this to happen we decided to take the players with the top 5 average minutes played in the season as they were likely the starting players for that team during that season. In order to do this, we had to engineer a new feature called average minutes played which would be the basis for our roster selections. 
    </p>

```python
updatedPlayerData['AVG Minutes Played'] = updatedPlayerData['gp'] * updatedPlayerData['usg_pct']
```

```python
# Filter the df to only include the max 5 of games played from each seasonEncoded and Team
updatedPlayerData = updatedPlayerData.groupby(['team', 'season']).apply(lambda x: x.nlargest(5, 'AVG Minutes Played')).reset_index(drop=True)

updatedPlayerData.head()
```

<div class="content">
    <h2>Train the ANN Model</h2>
    <p>
        To train our model to predict win rates based on team data we needed to do the following:
        <ul>
            <li> The data is first split into training and test sets using train_test_split, with 80% of the data used for training and 20% for testing. 
            <li> The features are then standardized using StandardScaler, ensuring that each feature has a mean of 0 and a standard deviation of 1 to improve model performance. 
            <li>The model itself is a sequential neural network consisting of three hidden layers with 64, 32, and 16 neurons respectively, each using the ReLU activation function. The output layer has a single neuron, suitable for regression, as the task is to predict a continuous value (win percentage). 
            <li>The model is compiled using the Adam optimizer and mean squared error (MSE) loss function, with mean absolute error (MAE) as a metric for evaluation. 
            <li>Additionally, early stopping is implemented to halt training if the validation loss stops improving after 10 epochs, restoring the best model weights to avoid overfitting. The model is trained for up to 100 epochs with a batch size of 5, and 20% of the training data is used for validation during training.
    </p>
</div>



```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(teamData, winRate, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test_Scaled = scaler.transform(X_test)
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)),  # Input layer
    keras.layers.Dense(64, activation='relu'),       # Hidden layer
    keras.layers.Dense(32, activation='relu'),       # Hidden layer
    keras.layers.Dense(16, activation='relu'),       # Hidden layer
    keras.layers.Dense(1)                             # Output layer (for regression)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(X_train, y_train, epochs=100, batch_size=5, validation_split=0.2, verbose=1, callbacks=[early_stopping])
```

    <h2>Test the ANN Model</h2>
    <p>
        This code evaluates the model's performance on the test set by comparing the predicted win rates to the actual win rates. It calculates the percentage of predictions that fall within 5 and 10 games of the actual win rate, showing how close the model’s predictions are to reality. Specifically, 39.3% of the predictions were within 5 games, and 60.7% were within 10 games. Additionally, common regression metrics are calculated: Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared (R²). The MAE of 0.10 reflects an average prediction error of about 8.2 games, the MSE of 0.02 captures overall error magnitude, and the R² of 0.26 indicates that the model explains 26% of the variance in win rates. These results highlight that while the model provides reasonable predictions, there is still significant room for improvement in accuracy and explanatory power.
    </p>
</div>



```python
# Example input for a new set of NBA players

predicted_win_rates = model.predict(X_test_Scaled)
actual_win_rates = y_test.values

# Values to keep track of how many predictions are within 5 and 10 games of the actual win rate
total = 0
within5 = 0
within10 = 0

# Display predictions alongside actual values
for pred, actual in zip(predicted_win_rates.flatten(), actual_win_rates):
    #print(f'Predicted: {pred:.3f}, Actual: {actual:.3f}')
    total += 1

    if abs(pred - actual) <= 5/82:
        within5 += 1
        within10 += 1
    elif abs(pred - actual) <= 10/82:
        within10 += 1

# print total, within5, within10
print(f'Total: {total}')
print(f'Within 5 Games: {within5/total * 100}%')
print(f'Within 10 Games: {within10/total * 100}%')
```
