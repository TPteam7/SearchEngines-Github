```python
import random
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# read in all the images from dogs
size_dogs = len(os.listdir('data/dogs'))
size_other = len(os.listdir('data/other'))

size_all = size_dogs + size_other

all_images = np.empty((size_all, 128, 128, 3), dtype=np.float32)      # 32 takes up less memory than 64

for i, image in enumerate(os.listdir('data/dogs')):
    img = cv.imread('data/dogs/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    img = cv.resize(img, (128, 128))
    img = img / 255.0

    # add the image to all_images
    all_images[i] = img

all_images.shape
```

```python
filtered_remaining_dogs = size_other     # filter to length of other images

np.random.shuffle(all_images)

random_dog_images = all_images[:filtered_remaining_dogs]
dog_labels = np.zeros((size_other, 1), dtype=np.float32)

print(len(random_dog_images))
```

```python
#size = len(os.listdir('data/other'))

#other_images = np.empty((size, 128, 128, 3), dtype=np.float32)      

for i, image in enumerate(os.listdir('data/other')):
    img = cv.imread('data/other/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    img = cv.resize(img, (128, 128))
    img = img / 255.0
    # add the image to all_images
    all_images[i + size_dogs] = img

all_images.shape
other_labels = np.ones((size_other, 1), dtype=np.float32)
```

```python
# combine the two labels
labels = np.concatenate((dog_labels, other_labels), axis=0)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    image = all_images[i]
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title(labels[i])
    plt.axis('off')
```

```python
# find any duplicates
duplicates = []
for i, img in enumerate(other_images):
    if np.array_equal(img, random_dog_images[0]):
        duplicates.append(i)

print(duplicates)
```

```python
# shuffle the other images
np.random.shuffle(other_images)

# ad the 0 labels to the other images
other_images_labels = [0] * len(other_images)
other_images_labels = np.array(other_images_labels)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    image = other_images[i]
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title("not dog")
    plt.axis('off')
```

```python
all_images = np.concatenate((random_dog_images, other_images), axis=0)
all_labels = np.concatenate((dog_labels, other_images_labels), axis=0)
```

```python
# make a list of indices
indices = list(range(len(all_images)))

# shuffle the indices
np.random.shuffle(indices)

# assign back those now shuffled indices
all_images = all_images[indices]
all_labels = all_labels[indices]
```

```python
all_images_size = len(random_dog_images) + len(other_images)

train_size = int(all_images_size * .8)
val_size = int(all_images_size * .1)
test_size = int(all_images_size * .1)

# Split the dataset into train, validation, and test. 80% train, 10% validation, 10% test
train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size).take(val_size)
test_dataset = dataset.skip(train_size + val_size).take(test_size)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNetV2

# Load pre-trained MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)  # Additional dense layer
output = Dense(1, activation='sigmoid')(x)   # binary classification

model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=5,              # Number of epochs to wait for improvement
    restore_best_weights=True # Restore weights of the best epoch
)

# print(X_train, X_train.shape)
model.fit(
    train_dataset, 
    validation_data = val_dataset, 
    epochs=50,
    callbacks= early_stopping)
```
