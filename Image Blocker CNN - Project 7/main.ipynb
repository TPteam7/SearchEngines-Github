```python
import random
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# read in all the images from dogs
all_dog_images = []

for image in os.listdir('data/dogs'):
    img = cv.imread('data/dogs/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    # add the image to all_images
    all_dog_images.append(img)

len(all_dog_images)

all_dog_images = np.array(all_dog_images)
```

```python
# read in all other images
other_images = []

for image in os.listdir('data/other'):
    img = cv.imread('data/other/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    # add the image to all_images
    other_images.append(img)

len(other_images)

other_images = np.array(other_images)
```

```python
filtered_remaining_dogs = len(other_images)

# randomly select the same number of images from the dog images
import random

random_dog_images = random.sample(all_dog_images, filtered_remaining_dogs)

len(random_dog_images)
```

```python
dog_labels = [1] * len(random_dog_images)

# shuffle the dog images
random.shuffle(random_dog_images)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    image = random_dog_images[i]
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title("dog")
    plt.axis('off')
```

```python
other_images_labels = [0] * len(other_images)

random.shuffle(other_images)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    image = other_images[i]
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title("not dog")
    plt.axis('off')
```

```python
# plot distribution of labels only dogs and not dogs
plt.hist(dog_labels + other_images_labels, bins=3)
plt.xticks([0, 1])
plt.title("Distribution of Labels")
plt.show()
```

potentailly need augmentation?

```python
random_dog_images = np.array(random_dog_images)
other_images = np.array(other_images)

random_dog_images = random_dog_images / 255.0
other_images = other_images / 255.0
```

```python
all_images = random_dog_images + other_images
all_labels = dog_labels + other_images

# Create a tf.data.Dataset
dataset = tf.data.Dataset.from_tensor_slices((all_images, all_labels))

# Shuffle and batch
dataset = dataset.shuffle(buffer_size=1000).batch(32).prefetch(tf.data.AUTOTUNE)
```
