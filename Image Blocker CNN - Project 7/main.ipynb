    <h2>Convolutional Neural Network (CNN) Algorithm</h2>
    <ol>
        <li>Initialize network architecture:
            <ul>
                <li>Define convolutional layers with randomly initialized filters/kernels</li>
                <li>Set up pooling layers for feature reduction</li>
                <li>Configure fully connected layers</li>
                <li>Choose appropriate activation functions (e.g., ReLU)</li>
            </ul>
        </li>
        <li><strong>Input Processing:</strong> Prepare input image data:
            <ul>
                <li>Normalize pixel values (typically to range 0-1 or -1 to 1)</li>
                <li>Resize images to consistent dimensions</li>
                <li>Preprocess images to standardize input format</li>
            </ul>
        </li>
        <li><strong>Convolutional Layer Operation:</strong> Apply convolution operations:
            <ul>
                <li>Slide convolutional filters across the input image</li>
                <li>Compute element-wise multiplications and summations</li>
                <li>Generate feature maps highlighting specific image characteristics</li>
                <li>Apply activation function (e.g., ReLU) to introduce non-linearity</li>
            </ul>
        </li>
        <li><strong>Pooling Layer:</strong> Reduce spatial dimensions:
            <ul>
                <li>Use max pooling or average pooling operations</li>
                <li>Downsample feature maps to extract most important features</li>
                <li>Reduce computational complexity</li>
                <li>Provide translation invariance</li>
            </ul>
        </li>
        <li><strong>Flattening:</strong> Prepare for fully connected layers:
            <ul>
                <li>Transform multi-dimensional feature maps into a 1D vector</li>
                <li>Prepare input for subsequent fully connected layers</li>
            </ul>
        </li>
        <li><strong>Fully Connected Layers:</strong> Final classification:
            <ul>
                <li>Apply dense layer connections</li>
                <li>Compute weighted sums with learned weights and biases</li>
                <li>Apply final activation function (e.g., softmax for multi-class classification)</li>
            </ul>
        </li>
        <li><strong>Loss Calculation:</strong> Evaluate network performance:
            <ul>
                <li>Compute loss using appropriate loss function (e.g., cross-entropy)</li>
                <li>Compare predicted outputs with true labels</li>
            </ul>
        </li>
        <li><strong>Backpropagation:</strong> Update network parameters:
            <ul>
                <li>Compute gradients for all layers using chain rule</li>
                <li>Update convolutional filters, weights, and biases</li>
                <li>Adjust parameters using optimization algorithm (e.g., Adam, SGD)</li>
            </ul>
            <p style="text-align: center;">
                <code>θ ← θ - η ∂L/∂θ</code>
            </p>
            <p>where <code>θ</code> represents network parameters, <code>η</code> is the learning rate, and <code>L</code> is the loss function.</p>
        </li>
        <li>Repeat training process for multiple epochs:
            <ul>
                <li>Iterate through entire training dataset</li>
                <li>Minimize loss and improve classification accuracy</li>
                <li>Implement techniques like early stopping to prevent overfitting</li>
            </ul>
        </li>
        <li>Validate and test the trained CNN:
            <ul>
                <li>Evaluate performance on separate validation dataset</li>
                <li>Make predictions on new, unseen images</li>
            </ul>
        </li>
    </ol>
</div>

```python
import random
import pandas as pd
import numpy as np
import tensorflow as tf
```

Get the dogs

```python
# read in all the images from dogs
size_dogs = len(os.listdir('data/dogs'))
size_other = len(os.listdir('data/other'))

size_all = size_other * 2       # do this since we are making the size dogs equally distributed to 

all_images = np.empty((size_all, 128, 128, 3), dtype=np.float32)      # 32 takes up less memory than 64

for i, image in enumerate(os.listdir('data/dogs')):
    if i > size_other:
        break
    img = cv.imread('data/dogs/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    img = cv.resize(img, (128, 128))
    img = img / 255.0

    # add the image to all_images
    all_images[i] = img

all_images.shape
```

Shuffle all images up to size other then make labels for them all

```python
np.random.shuffle(all_images[:size_other])
```

```python
dog_labels = np.ones((size_other), dtype=np.float32)
```

Get all the other images

```python
#size = len(os.listdir('data/other'))

#other_images = np.empty((size, 128, 128, 3), dtype=np.float32)      

for i, image in enumerate(os.listdir('data/other')):
    img = cv.imread('data/other/' + image)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    img = cv.resize(img, (128, 128))
    img = img / 255.0
    # add the image to all_images
    all_images[i + size_other] = img

all_images.shape
```

Make the 0 labels for all the other animals and concat them both

```python
other_labels = np.zeros((size_other), dtype=np.float32)
```

```python
# combine the two labels
labels = np.concatenate((dog_labels, other_labels), axis=0)
```

Shuffle the images with the labels attached

```python
combined = list(zip(all_images, labels))

np.random.shuffle(combined)
```

```python
all_images_shuffled, labels_shuffled = zip(*combined)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNetV2

# Load pre-trained MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)  # Additional dense layer
output = Dense(1, activation='sigmoid')(x)   # binary classification

model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# model.summary()
```
