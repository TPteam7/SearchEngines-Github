```python
import gzip
import json
import pandas as pd

dataList = []
current_line = 0

# Open and read the .json.gz file line by line until the end
with gzip.open('Sports_and_Outdoors_5.json.gz', 'rt', encoding='utf-8') as f:
    while True:
        line = f.readline()
        
        # Break the loop if readline() returns an empty string
        if not line:
            break

        try:
            # debug at every 250k lines
            if current_line % 250000 == 0:
                print(f"{current_line} lines reached.")

            # Parse JSON and append to the list
            data = json.loads(line)
            dataList.append(data)
            
            # Increment the line counter
            current_line += 1

        except json.JSONDecodeError as e:
            print(f"Error decoding JSON at line {current_line}: {e}")

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(dataList)
```

Keep only the important rating and text

```python
# only keep columns of overall and reviewText where verified is True
df = df[df['verified'] == True]
df = df[['overall', 'reviewText']]      # maybe keep summary to look at later

df.tail()
```

Drop null values and return the count

```python
df.isnull().sum()

# drop rows with missing values
df = df.dropna()

# check for missing values
df.isnull().sum()
```

Show the distribution of ratings

```python
# plot the distribution of the overall ratings
df['overall'].value_counts().sort_index().plot(kind='bar')

# print the counts of each rating
df['overall'].value_counts().sort_index()
```

Keep all ratings besides the 3 at around 100k. This will be used for feature engineering

```python
# i can only keep 100 k rows for both 4 and 5 ratings. make it random ones pulled
df_4 = df[df['overall'] == 4].sample(100000, random_state=42)
df_5 = df[df['overall'] == 5].sample(100000, random_state=42)

# drop all rows with 4 and 5 ratings from df
df = df.drop(df[(df['overall'] == 4) | (df['overall'] == 5)].index)

# combine the three dataframes
df_filtered = pd.concat([df, df_4, df_5])

# plot the distribution of the overall ratings
df_filtered['overall'].value_counts().sort_index().plot(kind='bar')

# print the counts of each rating
df_filtered['overall'].value_counts().sort_index()
```

#### **Feature Engineering**
-1 is negative and is 1 and 2 stars

0 is neutral and is 3 stars

1 is positive and is 4 and 5 stars

```python
# feauture engineering of -1 for 1 and 2 ratings, 0 for 3 ratings, 1 for 4 and 5 ratings
df_filtered['sentiment'] = df_filtered['overall'].apply(lambda x: -1 if x < 3 else 0 if x == 3 else 1)

# plot the distribution of the sentiment
df_filtered['sentiment'].value_counts().sort_index().plot(kind='bar')
```

```python
# reset the index
df_filtered = df_filtered.reset_index(drop=True)
```

```python
# remove the punctuation
df_filtered['reviewText'] = df_filtered['reviewText'].str.replace('[^\w\s]', '', regex=True)

# remove the new line characters
df_filtered['reviewText'] = df_filtered['reviewText'].str.replace('\n', '', regex=True)

# convert to lowercase
df_filtered['reviewText'] = df_filtered['reviewText'].str.lower()

# remove the extra whitespaces
df_filtered['reviewText'] = df_filtered['reviewText'].str.replace(' +', ' ', regex=True)

df_filtered['reviewText'].head()
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# remove the stopwords
df_filtered['reviewText'] = df_filtered['reviewText'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

# strip the whitespaces
df_filtered['reviewText'] = df_filtered['reviewText'].str.strip()

df_filtered['reviewText'].head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download the WordNet resource if you haven't already
nltk.download('wordnet')
nltk.download('punkt')  # Required for tokenization
```

```python
def lemmatize_text_with_pos(text):
    lemmatizer = WordNetLemmatizer()
    # Tokenize the input text
    tokens = nltk.word_tokenize(text)
    
    # Get POS tags
    pos_tags = nltk.pos_tag(tokens)
    
    # Lemmatize each token with its POS tag
    lemmatized_tokens = [
        lemmatizer.lemmatize(token, pos=get_wordnet_pos(pos)) for token, pos in pos_tags
    ]
    
    # Join the tokens back into a single string
    lemmatized_text = ' '.join(lemmatized_tokens)
    
    return lemmatized_text

def get_wordnet_pos(treebank_tag):
    """Convert treebank tags to WordNet POS tags."""
    if treebank_tag.startswith('J'):
        return 'a'  # Adjective
    elif treebank_tag.startswith('V'):
        return 'v'  # Verb
    elif treebank_tag.startswith('N'):
        return 'n'  # Noun
    elif treebank_tag.startswith('R'):
        return 'r'  # Adverb
    else:
        return 'n'  # Default to noun


df['reviewText'] = df['reviewText'].apply(lemmatize_text_with_pos)
```

```python
from nltk import word_tokenize, pos_tag

# Apply word tokenization to each text entry in the 'reviewText' column
df_filtered['tokenized'] = df_filtered['reviewText'].apply(lambda text: word_tokenize(str(text)))

# Apply POS tagging to each tokenized text
df_filtered['pos_tagged'] = df_filtered['tokenized'].apply(lambda tokens: pos_tag(tokens))

df_filtered['sentiment_words'] = df_filtered['pos_tagged'].apply(
    lambda tagged_words: [word for word, pos in tagged_words if pos in ['JJ', 'RB', 'VB', 'NN', 'NNS', 'NNP', 'NNPS']]
)

# View the resulting DataFrame
df_filtered[['reviewText', 'sentiment_words']]
```

```python
# View the resulting DataFrame
df_filtered[['sentiment', 'sentiment_words']]
```

```python
# Filter out rows where 'sentiment_words' is an empty list
df_filtered = df_filtered[df_filtered['sentiment_words'].apply(lambda x: len(x) > 2)]

# Now proceed with the rest of your processing (e.g., joining words and applying TfidfVectorizer)
df_filtered['sentiment_words_str'] = df_filtered['sentiment_words'].apply(lambda words: ' '.join(words))
```

```python
df_filtered.reset_index(drop=True, inplace=True)
```

```python
df_filtered["sentiment"]
```

```python
# Get the indices of rows with sentiment equal to 0
indices_to_drop = df_filtered[df_filtered['sentiment'] == 0].index

# Drop those indices from the DataFrame
df_filtered.drop(indices_to_drop, inplace=True)
```

```python
# After dropping the rows with sentiment value 0
df_filtered = df_filtered.reset_index(drop=True)
```

```python
score_baseline = df_filtered['sentiment'].value_counts(normalize=True).max()
score_baseline
```

```python
from sklearn.model_selection import train_test_split

X_train_text, X_test_text, y_train, y_test = train_test_split(df_filtered["sentiment_words_str"], df_filtered["sentiment"], test_size=0.8, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create tf-idf vectorizer
tfidf = TfidfVectorizer(max_features=2000)

# fit the vectorizer
X_train = tfidf.fit_transform(X_train_text)
```

```python
# transform the text
X_test = tfidf.fit_transform(X_test_text)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['Negative', 'Positive'])

print(f'Accuracy: {accuracy:.4f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
