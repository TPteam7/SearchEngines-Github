    <h2>Build the Model</h2>
    <p>
        Requirements:
    </p>
    <ul>
        <li>Remove punctuation</li>
        <li>Remove stop words (i.e., words that do not add a sentiment)</li>
        <li>Assign each word in every text element, with a sentiment score (use TfidVectorizer)</li>
        <li>Use a binary classification algorithm (e.g., logistic regression), which you can import from sklearn</li>
        <li>Divide the data into a training set and testing set, with a ratio of 80:20</li>
        <li>Fit the data set using the model</li>
        <li>Compute the (accuracy) score of the model</li>
    </ul>
</div>
