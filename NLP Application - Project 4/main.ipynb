```python
import matplotlib.pyplot as plt

def show_distribution(df, column, type):
    # plot the distribution of the column
    df[column].value_counts().sort_index().plot(kind='bar')

    # add a title to the plot
    plt.title(type)

    # show the plot
    plt.show()

show_distribution(testDf, 'label', 'Test')
show_distribution(trainDf, 'label', 'Train')
show_distribution(validDf, 'label', 'Valid')
```

```python
def clean_text(text):
    # remove the punctuation
    text = text.str.replace('[^\w\s]', '', regex=True)

    # remove the new line characters
    text = text.str.replace('\n', '', regex=True)

    # remove br tags
    text = text.str.replace('br', '', regex=True)

    # convert to lowercase
    text = text.str.lower()

    # remove the extra whitespaces
    text = text.str.replace(' +', ' ', regex=True)

    return text

testDf['text'] = clean_text(testDf['text'])
trainDf['text'] = clean_text(trainDf['text'])
validDf['text'] = clean_text(validDf['text'])

# print the 4647 index text
print(testDf['text'][4647])
print(testDf['label'][4647])
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop_words])

testDf['text'] = testDf['text'].apply(remove_stopwords)
trainDf['text'] = trainDf['text'].apply(remove_stopwords)
validDf['text'] = validDf['text'].apply(remove_stopwords)

# print the 4647 index text
print(testDf['text'][4647])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download the WordNet resource if you haven't already
nltk.download('wordnet')
nltk.download('punkt')  # Required for tokenization
```

```python
# lemmaization with pos tagging
from nltk.corpus import wordnet
from nltk import pos_tag

def lemmatize_text_with_pos(text):
    lemmatizer = WordNetLemmatizer()
    # Tokenize the input text
    tokens = nltk.word_tokenize(text)
    
    # Get POS tags
    pos_tags = nltk.pos_tag(tokens)
    
    # Lemmatize each token with its POS tag
    lemmatized_tokens = [
        lemmatizer.lemmatize(token, pos=get_wordnet_pos(pos)) for token, pos in pos_tags
    ]
    
    # Join the tokens back into a single string
    lemmatized_text = ' '.join(lemmatized_tokens)
    
    return lemmatized_text

def get_wordnet_pos(treebank_tag):
    """Convert treebank tags to WordNet POS tags."""
    if treebank_tag.startswith('J'):
        return 'a'  # Adjective
    elif treebank_tag.startswith('V'):
        return 'v'  # Verb
    elif treebank_tag.startswith('N'):
        return 'n'  # Noun
    elif treebank_tag.startswith('R'):
        return 'r'  # Adverb
    else:
        return 'n'  # Default to noun


testDf['text'] = testDf['text'].apply(lemmatize_text_with_pos)
trainDf['text'] = trainDf['text'].apply(lemmatize_text_with_pos)
validDf['text'] = validDf['text'].apply(lemmatize_text_with_pos)

# print the 4647 index text
print(testDf['text'][4647])
```
