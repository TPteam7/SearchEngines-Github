```python
import matplotlib.pyplot as plt

def show_distribution(df, column, type):
    # plot the distribution of the column
    df[column].value_counts().sort_index().plot(kind='bar')

    # add a title to the plot
    plt.title(type)

    # show the plot
    plt.show()

show_distribution(testDf, 'label', 'Test')
show_distribution(trainDf, 'label', 'Train')
```

```python
def clean_text(text):
    # remove the punctuation
    text = text.str.replace('[^\w\s]', '', regex=True)

    # remove the new line characters
    text = text.str.replace('\n', '', regex=True)

    # remove br tags
    text = text.str.replace('br', '', regex=True)

    # convert to lowercase
    text = text.str.lower()

    # remove the extra whitespaces
    text = text.str.replace(' +', ' ', regex=True)

    return text

testDf['text'] = clean_text(testDf['text'])
trainDf['text'] = clean_text(trainDf['text'])

# print the 4647 index text
print(testDf['text'][4647])
print(testDf['label'][4647])
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop_words])

testDf['text'] = testDf['text'].apply(remove_stopwords)
trainDf['text'] = trainDf['text'].apply(remove_stopwords)

# print the 4647 index text
print(testDf['text'][4647])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download the WordNet resource if you haven't already
nltk.download('wordnet')
nltk.download('punkt')  # Required for tokenization
```

```python
# lemmaization with pos tagging
from nltk.corpus import wordnet
from nltk import pos_tag

def lemmatize_text_with_pos(text):
    lemmatizer = WordNetLemmatizer()
    # Tokenize the input text
    tokens = nltk.word_tokenize(text)
    
    # Get POS tags
    pos_tags = nltk.pos_tag(tokens)
    
    # Lemmatize each token with its POS tag
    lemmatized_tokens = [
        lemmatizer.lemmatize(token, pos=get_wordnet_pos(pos)) for token, pos in pos_tags
    ]
    
    # Join the tokens back into a single string
    lemmatized_text = ' '.join(lemmatized_tokens)
    
    return lemmatized_text

def get_wordnet_pos(treebank_tag):
    """Convert treebank tags to WordNet POS tags."""
    if treebank_tag.startswith('J'):
        return 'a'  # Adjective
    elif treebank_tag.startswith('V'):
        return 'v'  # Verb
    elif treebank_tag.startswith('N'):
        return 'n'  # Noun
    elif treebank_tag.startswith('R'):
        return 'r'  # Adverb
    else:
        return 'n'  # Default to noun


testDf['text'] = testDf['text'].apply(lemmatize_text_with_pos)
trainDf['text'] = trainDf['text'].apply(lemmatize_text_with_pos)

# print the 4647 index text
print(testDf['text'][4647])
```

    <h2>Build the Model</h2>
    <p>
        Requirements:
    </p>
    <ul>
        <li>Assign each word in every text element, with a sentiment score (use TfidVectorizer)</li>
        <li>Use a binary classification algorithm (e.g., logistic regression), which you can import from sklearn</li>
        <li>Divide the data into a training set and testing set, with a ratio of 80:20</li>
        <li>Fit the data set using the model</li>
        <li>Compute the (accuracy) score of the model</li>
    </ul>
    <p>
        TF-IDF
    </p>
    <ul>
        <li>Text data is inherently unstructured and cannot be directly used by machine learning models. For the model to understand the data, we need to transform the text into a numerical format. The TF-IDF (Term Frequency-Inverse Document Frequency) Vectorizer is a common technique for this purpose.</li>
        <li>TF (Term Frequency) measures how frequently a word appears in a document.</li>
        <li>IDF (Inverse Document Frequency) reduces the importance of words that are common across the entire corpus. Words that are more unique to individual reviews will get higher weights, while overly common words get reduced importance.</li>
    </ul>
    <p>
        To maintain computational efficiency and reduce noise, we limit the vectorizer to the top 2000 features (words) based on their TF-IDF scores. This ensures that only the most important words for sentiment classification are retained.
    </p>
    <p>
        Once the vectorizer has been fitted on the training data, we can then transform the same data into a matrix of TF-IDF scores. This numerical matrix is what we will feed into the logistic regression model.
    </p>
</div>



```python
X_train = trainDf['text']
y_train = trainDf['label']

X_test = testDf['text']
y_test = testDf['label']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create tf-idf vectorizer
tfidf = TfidfVectorizer(max_features=2000)

# fit the vectorizer
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['Negative', 'Positive'])

print(f'Accuracy: {accuracy:.4f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
# test the model with a custom review
reviewText = "The movie was a waste of time"
review = tfidf.transform([reviewText])

print(f"Review: {reviewText}\nSentiment Score: {model.predict(review)}")

reviewText = "The movie was a great experience"
review = tfidf.transform([reviewText])

print(f"Review: {reviewText}\nSentiment Score: {model.predict(review)}")
```
